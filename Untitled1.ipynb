```python
import pandas as pd
import numpy as np
from datetime import datetime

print("Libraries loaded!")
print(f"Report generated at: {datetime.now().strftime('%Y-%m-%d %H:%M')}")
```

```python
# Şirkətdən gələn raw data (real həyatda CSV-dən oxunur)
data = {
    'Employee_ID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR', 'IT', 'Finance', 'HR'],
    'Hours_Worked': [40, 38, None, 42, 35, 40, None, 39],
    'Tasks_Completed': [12, 8, 15, 10, 7, 11, 9, None],
    'Performance_Score': [85, 72, 90, 88, 65, 78, 92, 70]
}

df = pd.DataFrame(data)
print("Raw data loaded:")
print(df)
```

```python
print("=" * 50)
print("DATA VALIDATION REPORT")
print("=" * 50)

# Boş dəyərləri tap
missing = df.isnull().sum()
print("\nMissing values found:")
print(missing[missing > 0])

# Duplicate-ları yoxla
duplicates = df.duplicated().sum()
print(f"\nDuplicate rows: {duplicates}")

# Data keyfiyyəti skoru
total_cells = df.shape[0] * df.shape[1]
missing_cells = df.isnull().sum().sum()
quality_score = ((total_cells - missing_cells) / total_cells) * 100
print(f"\nData Quality Score: {quality_score:.1f}%")
```

```python
# Missing dəyərləri ortalama ilə doldur
df['Hours_Worked'].fillna(df['Hours_Worked'].mean(), inplace=True)
df['Tasks_Completed'].fillna(df['Tasks_Completed'].mean(), inplace=True)

print("Missing values fixed!")
print(df)
```

```python
print("=" * 50)
print("AUTOMATED DEPARTMENT SUMMARY REPORT")
print(f"Generated: {datetime.now().strftime('%Y-%m-%d %H:%M')}")
print("=" * 50)

summary = df.groupby('Department').agg(
    Total_Employees=('Employee_ID', 'count'),
    Avg_Hours=('Hours_Worked', 'mean'),
    Avg_Tasks=('Tasks_Completed', 'mean'),
    Avg_Performance=('Performance_Score', 'mean')
).round(1)

print("\n", summary)

# Ən yaxşı department
best_dept = summary['Avg_Performance'].idxmax()
print(f"\n🏆 Best performing department: {best_dept}")
```

```python
# Report-u CSV-ə export et
timestamp = datetime.now().strftime('%Y%m%d_%H%M')
filename = f'department_report_{timestamp}.csv'

summary.to_csv(filename)
print(f"✅ Report saved as: {filename}")

# Cleaned data-nı da saxla
df.to_csv('cleaned_data.csv', index=False)
print("✅ Cleaned data saved!")
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Chart 1: Performance by Department
summary['Avg_Performance'].plot(kind='bar', ax=axes[0],
    color=['steelblue', 'salmon', 'green'])
axes[0].set_title('Avg Performance by Department')
axes[0].set_ylabel('Score')
axes[0].tick_params(axis='x', rotation=0)

# Chart 2: Hours Worked
summary['Avg_Hours'].plot(kind='bar', ax=axes[1],
    color=['steelblue', 'salmon', 'green'])
axes[1].set_title('Avg Hours Worked by Department')
axes[1].set_ylabel('Hours')
axes[1].tick_params(axis='x', rotation=0)

# Chart 3: Tasks Completed
summary['Avg_Tasks'].plot(kind='bar', ax=axes[2],
    color=['steelblue', 'salmon', 'green'])
axes[2].set_title('Avg Tasks Completed by Department')
axes[2].set_ylabel('Tasks')
axes[2].tick_params(axis='x', rotation=0)

plt.suptitle('Automated Department Performance Report',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('department_report.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Chart saved!")
```
